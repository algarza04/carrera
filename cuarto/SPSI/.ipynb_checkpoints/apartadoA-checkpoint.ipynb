El Cifrado Vigenère: Teoría y Contexto

Basándonos en la Sección 5.2 del libro An Introduction to Mathematical Cryptography, los cifrados de sustitución simple (como el César) son monoalfabéticos, ya que cada letra del texto plano se cifra siempre con el mismo alfabeto cifrado.

A medida que el criptoanálisis avanzó en la Italia del Renacimiento, surgió la necesidad de sistemas más seguros. El Cifrado Vigenère es un ejemplo de cifrado polialfabético.

    Nota Histórica: Este cifrado lleva el nombre de Blaise de Vigenère (1523–1596), cuyo libro de 1586 Traicté des Chiffres describía los cifrados de su época. Curiosamente, el libro menciona que Vigenère probablemente no inventó este cifrado específico, aunque sí inventó un ingenioso sistema de "autoclave".

En el libro Advances in Information Systems se añade que, aunque data del siglo XVI, sigue siendo un método práctico. Su fortaleza radica en que una misma letra del texto plano puede ser representada por diferentes letras en el texto cifrado dependiendo de su posición, lo que oculta la estructura del mensaje y dificulta el análisis de frecuencias simple.

El cifrado Vigenère es polialfabético. La fórmula matemática para cifrar el i-ésimo carácter del mensaje P con la clave K de longitud m es:
Ci​=(Pi​+Ki(mod m)​)(mod26)

```python
# ciframos con vigenere el texto a partir de la clave 
def cifrado_vigenere(texto, clave):
    texto = texto.upper().replace(" ", "")
    clave = clave.upper().replace(" ", "")
    resultado = ""
    
    for i, char in enumerate(texto):
        if char.isalpha():
            # Obtenemos el valor de la posición i de la clave
            clave_char = clave[i % len(clave)]
            
            # Convertimos el char del texto y de la clave a su correspondiente valor ascii y restamos
            # el valor de la primera posición(A) para tener los valores entre 0 y 25
            texto_num = ord(char) - ord('A')
            clave_num = ord(clave_char) - ord('A')
            
            # Aplicamos el cifrado
            cifrado_num = (texto_num + clave_num) % 26
            
            # Convertimos el número cifrado a char de nuevo y lo añadimos al resultado
            resultado += chr(cifrado_num + ord('A'))
    
    return resultado

# desciframos con vigener el texto a partir de la clave
def descifrado_vigenere(texto_cifrado, clave):
    texto_cifrado = texto_cifrado.upper().replace(" ", "")
    clave = clave.upper().replace(" ", "")
    resultado = ""
    
    for i, char in enumerate(texto_cifrado):
        if char.isalpha():
            # Obtenemos el valor de la posición i de la clave
            clave_char = clave[i % len(clave)]
            
            # Convertimos el char del texto y de la clave a su correspondiente valor ascii y restamos
            # el valor de la primera posición(A) para tener los valores entre 0 y 25            
            texto_num = ord(char) - ord('A')
            clave_num = ord(clave_char) - ord('A')
            
            # Aplicamos descifrado
            descifrado_num = (texto_num - clave_num) % 26
            
            # Convertimos el número cifrado a char de nuevo y lo añadimos al resultado
            resultado += chr(descifrado_num + ord('A'))
        else:
            resultado += char
    
    return resultado
```

Vamos a comprobar que esto funcione con el ejemplo que vemos en el apartado 5.2 de An Introduction to Mathematical Cryptography.

```python
# bsrocc
mensaje = "y,ellow"
clave = "dog"

# Ciframos el texto
cifrado = cifrado_vigenere(mensaje, clave)
print(f"Mensaje original: {mensaje}")
print(f"Clave:            {clave}")
print(f"Cifrado:          {cifrado}") 

# Desciframos el texto
descifrado = descifrado_vigenere(cifrado, clave)
print(f"Descifrado:       {descifrado}")
```

Implementación en python del criptosistema de vigenère:

```python
def cifrado_vigenere(texto, clave):
    """
    Cifra un texto usando el cifrado de Vigenère
    
    Args:
        texto: Texto a cifrar
        clave: Clave para el cifrado
    
    Returns:
        Texto cifrado
    """
    texto = texto.upper().replace(" ", "")
    clave = clave.upper().replace(" ", "")
    resultado = ""
    
    for i, char in enumerate(texto):
        if char.isalpha():
            # Obtener la letra de la clave correspondiente (repetir clave si es necesario)
            clave_char = clave[i % len(clave)]
            
            # Convertir a números (A=0, B=1, ..., Z=25)
            texto_num = ord(char) - ord('A')
            clave_num = ord(clave_char) - ord('A')
            
            # Aplicar el cifrado: (texto + clave) mod 26
            cifrado_num = (texto_num + clave_num) % 26
            
            # Convertir de vuelta a letra
            resultado += chr(cifrado_num + ord('A'))
        else:
            resultado += char
    
    return resultado


def descifrado_vigenere(texto_cifrado, clave):
    """
    Descifra un texto usando el cifrado de Vigenère
    
    Args:
        texto_cifrado: Texto cifrado
        clave: Clave usada en el cifrado
    
    Returns:
        Texto descifrado
    """
    texto_cifrado = texto_cifrado.upper().replace(" ", "")
    clave = clave.upper().replace(" ", "")
    resultado = ""
    
    for i, char in enumerate(texto_cifrado):
        if char.isalpha():
            # Obtener la letra de la clave correspondiente
            clave_char = clave[i % len(clave)]
            
            # Convertir a números
            texto_num = ord(char) - ord('A')
            clave_num = ord(clave_char) - ord('A')
            
            # Aplicar el descifrado: (texto - clave) mod 26
            descifrado_num = (texto_num - clave_num) % 26
            
            # Convertir de vuelta a letra
            resultado += chr(descifrado_num + ord('A'))
        else:
            resultado += char
    
    return resultado
```

## Ejemplos de uso

```python
# Ejemplo 1: Cifrado
texto_original = "HOLA MUNDO"
clave = "CLAVE"

texto_cifrado = cifrado_vigenere(texto_original, clave)
print(f"Texto original: {texto_original}")
print(f"Clave: {clave}")
print(f"Texto cifrado: {texto_cifrado}")

# Ejemplo 2: Descifrado
texto_descifrado = descifrado_vigenere(texto_cifrado, clave)
print(f"\nTexto descifrado: {texto_descifrado}")
```

```python
# Ejemplo 3: Mensaje más complejo
mensaje = "ATAQUE AL AMANECER"
clave_secreta = "SECRETO"

print(f"Mensaje original: {mensaje}")
print(f"Clave: {clave_secreta}")

cifrado = cifrado_vigenere(mensaje, clave_secreta)
print(f"Mensaje cifrado: {cifrado}")

descifrado = descifrado_vigenere(cifrado, clave_secreta)
print(f"Mensaje descifrado: {descifrado}")
```

```python
import math
from collections import Counter

def indice_coincidencia(texto):
    """
    Calcula el índice de coincidencia de un texto.
    IC = Σ(ni * (ni-1)) / (N * (N-1))
    donde ni es la frecuencia de cada letra y N es la longitud total
    """
    texto = texto.upper().replace(" ", "")
    N = len(texto)
    if N <= 1:
        return 0
    
    frecuencias = Counter(texto)
    suma = sum(f * (f - 1) for f in frecuencias.values())
    
    return suma / (N * (N - 1))


def longitud_clave_kasiski(texto_cifrado, max_longitud=20):
    """
    Estima la longitud de la clave usando el método de Kasiski.
    Busca secuencias repetidas y calcula el MCD de las distancias.
    """
    texto = texto_cifrado.upper().replace(" ", "")
    secuencias = {}
    
    # Buscar secuencias de 3 letras repetidas
    for longitud_seq in range(3, 6):
        for i in range(len(texto) - longitud_seq):
            seq = texto[i:i+longitud_seq]
            if seq in secuencias:
                secuencias[seq].append(i)
            else:
                secuencias[seq] = [i]
    
    # Calcular distancias entre repeticiones
    distancias = []
    for seq, posiciones in secuencias.items():
        if len(posiciones) > 1:
            for i in range(len(posiciones) - 1):
                distancias.append(posiciones[i+1] - posiciones[i])
    
    if not distancias:
        return None
    
    # Encontrar el MCD de todas las distancias
    def mcd(a, b):
        while b:
            a, b = b, a % b
        return a
    
    resultado = distancias[0]
    for d in distancias[1:]:
        resultado = mcd(resultado, d)
    
    # Contar factores comunes
    factores = Counter()
    for d in distancias:
        for i in range(2, min(d, max_longitud) + 1):
            if d % i == 0:
                factores[i] += 1
    
    # Retornar el factor más común
    if factores:
        return factores.most_common(1)[0][0]
    return resultado


def longitud_clave(texto_cifrado, max_longitud=20):
    """
    Estima la longitud de la clave usada en el cifrado de Vigenère
    usando el método del índice de coincidencia.
    
    Args:
        texto_cifrado: Texto cifrado
        max_longitud: Longitud máxima a probar
    
    Returns:
        Longitud estimada de la clave
    """
    texto = texto_cifrado.upper().replace(" ", "")
    IC_ESPERADO = 0.065  # IC para español/inglés
    
    mejores_longitudes = []
    
    # Probar diferentes longitudes
    for longitud in range(1, max_longitud + 1):
        # Dividir el texto en grupos según la longitud
        grupos = ['' for _ in range(longitud)]
        
        for i, char in enumerate(texto):
            if char.isalpha():
                grupos[i % longitud] += char
        
        # Calcular el IC promedio de todos los grupos
        ics = [indice_coincidencia(grupo) for grupo in grupos if len(grupo) > 1]
        
        if ics:
            ic_promedio = sum(ics) / len(ics)
            # Guardar longitud y qué tan cerca está del IC esperado
            diferencia = abs(ic_promedio - IC_ESPERADO)
            mejores_longitudes.append((longitud, ic_promedio, diferencia))
            print(f"Longitud {longitud:2d}: IC promedio = {ic_promedio:.4f}")
    
    # Ordenar por menor diferencia con IC esperado
    mejores_longitudes.sort(key=lambda x: x[2])
    
    print(f"\n⭐ Longitud más probable: {mejores_longitudes[0][0]} (IC = {mejores_longitudes[0][1]:.4f})")
    print(f"Otras posibilidades: {[l[0] for l in mejores_longitudes[1:4]]}")
    
    return mejores_longitudes[0][0]
```

```python
def encontrar_clave(texto_cifrado, longitud_clave):
    """
    Encuentra la clave usada en el cifrado de Vigenère mediante análisis de frecuencias.
    
    Args:
        texto_cifrado: Texto cifrado
        longitud_clave: Longitud de la clave
    
    Returns:
        Clave estimada
    """
    texto = texto_cifrado.upper().replace(" ", "")
    
    # Frecuencias esperadas en español (pueden ajustarse para inglés)
    FREQ_ESPANOL = {
        'E': 13.68, 'A': 12.53, 'O': 8.68, 'S': 7.98, 'R': 6.87,
        'N': 6.71, 'I': 6.25, 'D': 5.86, 'L': 4.97, 'C': 4.68,
        'T': 4.63, 'U': 3.93, 'M': 3.15, 'P': 2.51, 'B': 1.42,
        'G': 1.01, 'V': 0.90, 'Y': 0.90, 'Q': 0.88, 'H': 0.70,
        'F': 0.69, 'Z': 0.52, 'J': 0.44, 'Ñ': 0.31, 'X': 0.22,
        'W': 0.02, 'K': 0.01
    }
    
    clave = ""
    
    # Para cada posición de la clave
    for pos in range(longitud_clave):
        # Extraer todas las letras cifradas con esta posición de clave
        subcadena = ""
        for i in range(pos, len(texto), longitud_clave):
            if texto[i].isalpha():
                subcadena += texto[i]
        
        # Contar frecuencias en la subcadena
        frecuencias = Counter(subcadena)
        total = len(subcadena)
        
        # Probar cada posible letra de la clave (A-Z)
        mejor_letra = 'A'
        mejor_puntuacion = float('inf')
        
        for shift in range(26):
            # Descifrar subcadena con este shift
            puntuacion = 0
            for letra, count in frecuencias.items():
                # Aplicar descifrado
                letra_descifrada = chr(((ord(letra) - ord('A') - shift) % 26) + ord('A'))
                
                # Comparar con frecuencia esperada
                freq_observada = (count / total) * 100
                freq_esperada = FREQ_ESPANOL.get(letra_descifrada, 0.01)
                puntuacion += abs(freq_observada - freq_esperada)
            
            if puntuacion < mejor_puntuacion:
                mejor_puntuacion = puntuacion
                mejor_letra = chr(shift + ord('A'))
        
        clave += mejor_letra
        print(f"Posición {pos + 1}: '{mejor_letra}' (puntuación: {mejor_puntuacion:.2f})")
    
    return clave
```
